```python
import numpy as np
y = [0,3,7,4,6,9,8,3,2,1]
x = np.linspace(0, 24, len(y))
x = list(x)
y = list(y)
```

```python
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y, label='Ma première courbe !')
plt.show()
```

```python
plt.plot(x, y, color='red')
plt.xlabel(" L'axe des abscisses")
plt.ylabel("L'axe des ordonnées")
plt.title("Ma deuxième courbe !")
plt.show()
```

```python
x = list(np.linspace(0, 24, 10))

y1 = [0,3,7,4,6,9,8,3,2,1] # temperature
y2 = [2,8,1,2,1,4,7,8,6,7] # humidité

plt.plot(x, y1, color='blue')
plt.plot(x, y2, color='red')
plt.xlabel(" L'axe des abscisses")
plt.ylabel("L'axe des ordonnées")
plt.title("multi-courbes")
plt.show()


#ax.set_xlabel(" L'axe des abscisses")
#ax.set_ylabel("L'axe des ordonnées")
#ax.set_title("multi-courbes")

#plt.show() #c'est plt qui continue d'indiquer l'affichage du graphique
```

```python
fig = plt.figure() # je créais un nouvel espace de création pour mes graphique
ax1 = fig.add_subplot (221) #Je créais un nouveau graphique appelé 'ax' et j'indique où il se trouve
ax1.plot(x,y1, color='lightblue') # je ne dis plus plt mais ax pour indiqué ce qui est présent
ax1.plot(x,y2, color='darkgreen')


ax2 = fig.add_subplot(222) #Je créais un nouveau graphique appelé 'ax' et j'indique où il se trouve
ax2.plot(x,y1, color='red') # je ne dis plus plt mais ax pour indiqué ce qui est présent
ax2.plot(x,y2, color='blue')


ax3 = fig.add_subplot(223) #Je créais un nouveau graphique appelé 'ax' et j'indique où il se trouve
ax3.plot(x,y1, color='green') # je ne dis plus plt mais ax pour indiqué ce qui est présent
ax3.plot(x,y2, color='yellow')


ax4 = fig.add_subplot(224) #Je créais un nouveau graphique appelé 'ax' et j'indique où il se trouve
ax4.scatter(x,y1, color='red') # je ne dis plus plt mais ax pour indiqué ce qui est présent
ax4.scatter(x,y2, color='yellow')

plt.savefig("haythem.png")
plt.savefig("haythem.pdf")
plt.savefig("haythem.svg")
```
